```javascript
var csv = require('csv-parse')
var parse = require('csv-parse/lib/sync')
var fs = require('fs')

var raw = fs.readFileSync(__dirname + '/2015senatebills.csv')
raw = parse(raw, {delimiter: ','})

```

```javascript
raw = raw.slice(1)
```

```javascript
data = {}
raw.forEach((r)=>{
    
    const key = r[4].replace(' ', '') + ' ' + r[3]
    const value = {}
    value.title = r[7]
    value.tags = r[1]
    value.description = r[2].split('\n\n')[1].trim()
    value.date = r[6]
    value.progressivePosition = r[10].replace('+', '').trim()
    value.yesVotes = r[11]
    value.noVotes = r[12]
    
    let body = r[4][0] === 'S' ? 'Senate' : null
    body = r[4][0] === 'H' ? 'House' : r[4][0]
    value.url = `https://malegislature.gov/Bills/189/${key.replace('.', '').trim().split(' ')[0]}` 
    
    data[key] = value
    
})
```

```javascript
data
```

```javascript
fs.writeFile('./senate_bills.json', JSON.stringify(data), function(err) {
  if (err) {
    return console.log(err);
  }

  console.log("The file was saved!");
});
```
