```javascript
var csv = require('csv-parse')
var parse = require('csv-parse/lib/sync')
var fs = require('fs')
var turf = require('turf')

var mass = require('./SENATE2012.json')



```

```javascript
// IN FORM LONGITUDE,LATITUDE
point = [-71.1097, 42.3736]
```

```javascript
function findDistrict (longLat) {
    const point = turf.point(longLat)
    const matches = mass.features.filter((f)=>{
        const polygon = turf.multiPolygon(f.geometry.coordinates)
        return turf.inside(point, polygon)
    })
    if (matches.length){
        return matches[0].properties
    }
}
```

```javascript
district = findDistrict(point)
```
