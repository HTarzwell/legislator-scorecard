```javascript
var axios = require('axios')
var csv = require('csv-parse')
var parse = require('csv-parse/lib/sync')
var fs = require('fs')
var axios = require('axios')

var raw = fs.readFileSync(__dirname + '/2017/legislation-tracking.csv')
raw = parse(raw, {delimiter: ','})
```

```javascript
data = {}
raw.slice(1).forEach((row)=>{
    data[row[1].trim()] = {
        title : row[5].trim(),
        description : row[10].trim()
    }
})
```

```javascript
fs.writeFile('./2017/legislation-description.json', JSON.stringify(data), function(err) {
  if (err) {
    return console.log(err);
  }

  console.log("The file was saved!");
});
```
